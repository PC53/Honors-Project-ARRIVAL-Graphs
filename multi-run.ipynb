```python
def multi_run_procedure(instance: Arrival, size_of_S : int):
    s_curr =  instance.s_0
    s_next = instance.s_1
    
    S_subset = rm.sample(instance.vertices[1:-1],size_of_S)
    print("subset S is ",S_subset)
    #### TODO 
    w = {s:1 for s in S_subset}
    
    t = {s:0 for s in instance.vertices}
    t[0] = 1
    for v in S_subset:
        t[instance.s_0[v]] += np.ceil(w[v]/2)
        t[instance.s_1[v]] += np.floor(w[v]/2)
    # print(t)
    
    waiting_set = []
    for v in instance.vertices:
        if v not in S_subset and v != -1 and v != instance.n-1:
            waiting_set.append(v)
            
    while len(waiting_set)>0:
        waiting_set_ = [ws for ws in waiting_set if t[ws]>0]
        if not waiting_set_:
            break
        
        choose = rm.choice(waiting_set_)
        # print(choose,t)
        if t[choose] <= 0 :
            # waiting_set.remove(choose)
            print(waiting_set)
        else:
            tau = rm.randint(1,t[choose])
            t[choose] -= tau
            t[s_curr[choose]] += np.ceil(tau/2)
            t[s_next[choose]] += np.floor(tau/2)
            
            if tau & 1 : 
                temp = s_curr[choose]
                s_curr[choose] = s_next[choose]
                s_next[choose] = temp
            
        # break
    print("final t array ",t) 
    return t
```

```python
multi_run_procedure(instance, 10)
```
